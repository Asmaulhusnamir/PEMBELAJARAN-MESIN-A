```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/test (1).csv')
df1
```

```python
df2 = df1.dropna()
```

```python
df2
```

```python
df2.describe()
```

```python
import sklearn.model_selection as ms
x=df2[['x']]
y=df2[['y']]
X_train, X_test, y_train, y_test = ms.train_test_split(x,y,test_size=0.2, random_state=0)
print(X_train.size, X_test.size)
#20% menjadi test data dan 80%menjadi training data
```

**Visualisasi**

```python
import numpy as np
import matplotlib.pyplot as plt
plt.scatter(X_train,y_train, edgecolors='y')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot antara x dan y')
x1 = np.linspace(0,106)
y1 = -0.53332421 + 1.0144497*x1
plt.plot(x1,y1)
plt.show()
```

```python
import sklearn.linear_model as lm
model1 = lm.LinearRegression()
model1.fit(X_train,y_train)
```

```python
model1.coef_
```

```python
model1.intercept_
#KETIKA MELEWATI SUMBU Y, DIA BERPOTONGAN DI TITIK -0,5
```

```python
#Menguji model yang kita latih 
# bebas itu X, terikat itu Y
#r square menentukan seberap adekat model kita dengan garis
#r-square mendekati 1 itu bagus
#R^2 = 1-(RSS/TSS), data bagi garis linear
r2 = model1.score(X_test, y_test)
print(r2)
```

```python
#prediksi diatas bagus karen amendekati 1
jarak = model1.predict([[50]])
print(jarak)
```

##MENGGUBKANA LEBIH DARI 1 VARIABEL BEBA(MULTIPLE LINEAR REGRESSIONS)##

```python
df = pd.read_csv('/content/kc_house_data.csv')
```

```python
df
```

```python
#agar pembuatan model tidak sia-sia, maka asumsinya dibenarkan. mislnya kalau ada variabel yang tidak berpengaruh sama skli maka bagusnya dibuang saja kek mantan
df.corr(method='pearson')
```

```python
import pandas as pd
import numpy as np
import sklearn.model_selection as ms
import sklearn.linear_model as lm
X1 = df[['bathrooms', 'sqft_living']]
y1 = df[['price']]
X1_train, X1_test, y1_train, y1_test = ms.train_test_split(X1,y1,test_size=0.2, random_state=0)
print(X1_train.size, X1_test.size)
```

```python
model = lm.LinearRegression()
model.fit(X1_train, y1_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
data1 = np.array([[3,2000]])
price = model.predict(data1)
print(price)
```

```python
r2 = model.score(X1_test, y1_test)
print (r2)
```
